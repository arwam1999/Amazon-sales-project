```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import re#because i want use r" in writing 
from matplotlib.ticker import MultipleLocator
```

```python
df= pd.read_csv('amazon.csv')
pd.set_option('display.max_rows', None)# display all rows
df.head()
```

```python
df[['discounted_price', 'actual_price', 'discount_percentage','rating', 'rating_count']].head()
#^set a column in a data frame 
# dubble Qutation because if i want multi columns , single qutation one coulmn
```

```python
df.info()# datatype object
```

```python
#Must before convert correct data in columns it is must without symbols
def remove_symbols(x):
        # remove anything 
        Cleaned=re.sub(r'[^\d.]', '', str(x))
        return Cleaned if Cleaned!='' else None

#Apply functon on columns 
df['discounted_price']=df['discounted_price'].apply(remove_symbols)
df['actual_price']=df['actual_price'].apply(remove_symbols)
df['rating']=df['rating'].apply(remove_symbols)
df['rating_count']=df['rating_count'].apply(remove_symbols)
df['discount_percentage']=df['discount_percentage'].apply(remove_symbols)
```

```python
# Convert type of column to float 
df['discounted_price'] = df['discounted_price'].astype(float)
df['actual_price']=df['actual_price'].astype(float)
df['rating']=df['rating'].astype(float)
df['rating_count']=df['rating_count'].astype(float)
df['discount_percentage']=df['discount_percentage'].astype(float)

#This code is remove errors from cloumn becouse Nan some times it cause error
#df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
#df['actual_price']=pd.to_numeric(df['discounted_price'], errors='coerce')
```

```python
plt.hist(df['discounted_price'].dropna(), bins=10, color='lightgreen', edgecolor='black')
plt.xlabel('Discounted Price')
plt.ylabel('Number of Products')
plt.title('Distribution of Discounted Prices')
plt.show()
```

```python
# Calcultae averge ber rating 
avg_rating_per_product = df.groupby('product_name')['rating'].mean()

#The product has lowiest value
lowest_rated_product = avg_rating_per_product.idxmin()
lowest_rating = avg_rating_per_product.min()

print(f"The loweist value is: {lowest_rated_product}")
print(f"Average rating: {lowest_rating:.2f}")
```

```python
#Ctegorised discount pprecentage
def category_discount(x):
    if (x>0) & (x<25):
        return 'Low'
    elif (x>25) & (x<54):
            return 'Medium'
    elif (x>54) & (x<100):
            return'High'

df['category_discount']=df['discount_percentage'].apply(category_discount)
print(df[['category_discount', 'discount_percentage']])
```

```python
#It is effect discount precentage on products 

print(df['category_discount'].value_counts())
```

### Observation
It is efficet discount price on Number 

```python
#Arrange 
def Arrange_category(x):
    if str(x).startswith('Computers'):
        return 'Computers'
    elif str(x).startswith('Home'):
        return 'Home'
    elif str(x).startswith('OfficeProducts'):
        return 'Office'
    elif str(x).startswith('Electronics'):
        return 'Electronics'
    elif str(x).startswith('Musical'):
        return 'Electronics'
    else:
        return 'None'

df['Category_arrange']=df['category'].apply(Arrange_category)
```

```python
print(df['Category_arrange'].value_counts())
```

```python
#Discount precentage in every category
df.groupby('Category_arrange')['discount_percentage'].mean()
```
